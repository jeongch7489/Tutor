```python
def genre_count(df):
    word = input('enter the word').lower()
    clean_app = df['App'].apply(lambda x : re.sub(r'[^a-zA-Z0-9]', ' ', x.lower()).replace("  ", " ").split(' ') if type(x) == str else x)
    new_genre = df['Genres'].replace(';', ', ', regex=True)
    new_genre = new_genre.apply(lambda x: x.split(', ') if type(x) == str else x)
    genre = []
    for k, v in enumerate(clean_app):
        if word in v:
            genre.append(new_genre[k])
    genre = [x for xs in genre for x in xs]
    values, counts = np.unique(genre, return_counts=True)
    zipped = zip(values, counts)
    return dict(zipped)
            
genre_count(data)
```
